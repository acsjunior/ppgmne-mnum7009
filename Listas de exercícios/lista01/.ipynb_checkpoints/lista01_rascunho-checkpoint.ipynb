```python
import numpy as np
```

```python
# a)

A = np.array([[1/2, 1/3], [1/3, 1/4]])
b = np.array([[1/63], [1/168]])
x = np.array([[1/7, -1/6]]).T
x_ = np.array([[0.142, -0.166]]).T
```

```python
np.matmul(A, x)
```

```python
np.matmul(A, x_)
```

```python
x - x_
```

```python
np.matmul(A, x_) - b
```

```python
# Norma infinita:

np.linalg.norm(np.matmul(A, x_) - b, ord=np.infty)
```

```python
# Norma 1:

np.linalg.norm(np.matmul(A, x_) - b, ord=1)
```

```python
# Norma 2:

np.linalg.norm(np.matmul(A, x_) - b, ord=2)
```
