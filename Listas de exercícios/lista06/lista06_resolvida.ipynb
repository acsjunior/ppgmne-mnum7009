$L_0(x) = \dfrac{(x - x_1)(x - x_2)(x - x_3)}{(x_0 - x_1)(x_0 - x_2)(x_0 - x_3)}$

$L_0(x) = \dfrac{(x + 1)(x - 1)(x - 2)}{(-2 + 1)(-2 - 1)(-2 - 2)}$

$L_0(x) = \dfrac{(x + 1)(x - 1)(x - 2)}{(-1)(-3)(-4)}$

$L_0(x) = \dfrac{(x + 1)(x - 1)(x - 2)}{-12}$

$L_0(x) = \dfrac{x^3 - 2x^2 - x + 2}{-12}$

$L_0(x) = \frac{-1}{12}x^3 + \frac{1}{6}x^2 + \frac{1}{12}x - \frac{1}{6}$

<br>
<br>
<br>

$L_1(x) = \dfrac{(x - x_0)(x - x_2)(x - x_3)}{(x_1 - x_0)(x_1 - x_2)(x_1 - x_3)}$

$L_1(x) = \dfrac{(x + 2)(x - 1)(x - 2)}{(-1 + 2)(-1 - 1)(-1 - 2)}$

$L_2(x) = \dfrac{(x + 2)(x + 1)(x - 2)}{(1 + 2)(1 + 1)(1 - 2)}$

$L_2(x) = \dfrac{x^3 + x^2 - 4x - 4}{-6}$

```python
X = np.array([-2, -1, 1, 2])

def L0(x): return ((x-X[1])*(x-X[2])*(x-X[3])) / ((X[0]-X[1])*(X[0]-X[2])*(X[0]-X[3]))

def L1(x): return ((x-X[0])*(x-X[2])*(x-X[3])) / ((X[1]-X[0])*(X[1]-X[2])*(X[1]-X[3]))

def L2(x): return ((x-X[0])*(x-X[1])*(x-X[3])) / ((X[2]-X[0])*(X[2]-X[1])*(X[0]-X[3]))

def L3(x): return ((x-X[0])*(x-X[1])*(x-X[2])) / ((X[3]-X[0])*(X[3]-X[1])*(X[3]-X[2]))
```

```python
X = np.array([5.9, 6.0, 6.1, 6.2])
Y = np.array([34.8, 36.0, 37.2, 38.4])

def f0(): return Y[0]

def f01(): return (Y[1]-Y[0]) / (X[1]-X[0])

def f12(): return (Y[2]-Y[1])/(X[2]-X[1])

def f23(): return (Y[3]-Y[2])/(X[3]-X[2])

def f012(): return (f12() - f01()) / (X[2] - X[0])

def f123(): return (f23() - f12()) / (X[3] - X[1])

def f0123(): return (f123() - f012()) / (X[3] - X[0])

def P_newton3(x):
    d0 = f0()
    d1 = f01()
    d2 = f012()
    d3 = f0123()
    return d0 + d1*(x-X[0]) + d2*(x-X[0])*(x-X[1]) + d3*(x-X[0])*(x-X[1])*(x-X[2])
```

```python
X = np.array([1940, 1950, 1960, 1970, 1980, 1990])
Y = np.array([132165, 151326, 179323, 203302, 226542, 249633])
```

```python
def f0(): return Y[0]

def f01(): return (Y[1]-Y[0]) / (X[1]-X[0])
def f12(): return (Y[2]-Y[1])/(X[2]-X[1])
def f23(): return (Y[3]-Y[2])/(X[3]-X[2])
def f34(): return (Y[4]-Y[3])/(X[4]-X[3])
def f45(): return (Y[5]-Y[4])/(X[5]-X[4])

def f012(): return (f12() - f01()) / (X[2] - X[0])
def f123(): return (f23() - f12()) / (X[3] - X[1])
def f234(): return (f34() - f23()) / (X[4] - X[2])
def f345(): return (f45() - f34()) / (X[5] - X[3])

def f0123(): return (f123() - f012()) / (X[3] - X[0])
def f1234(): return (f234() - f123()) / (X[4] - X[1])
def f2345(): return (f345() - f234()) / (X[5] - X[2])

def f01234(): return (f1234() - f0123()) / (X[4] - X[0])
def f12345(): return (f2345() - f1234()) / (X[5] - X[1])

def f012345(): return (f12345() - f01234()) / (X[5] - X[0])

def P_newton5(x):
    d0 = f0()
    d1 = f01()
    d2 = f012()
    d3 = f0123()
    d4 = f01234()
    d5 = f012345()

    return (d0 + 
            d1*(x-X[0]) + 
            d2*(x-X[0])*(x-X[1]) + 
            d3*(x-X[0])*(x-X[1])*(x-X[2]) + 
            d4*(x-X[0])*(x-X[1])*(x-X[2])*(x-X[3]) + 
            d5*(x-X[0])*(x-X[1])*(x-X[2])*(x-X[3])*(x-X[4]))
```
